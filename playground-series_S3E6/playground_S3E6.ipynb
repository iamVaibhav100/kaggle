```python
import pandas as pd
import numpy as np
```

```python
X = pd.read_csv("sample_data/train.csv")
X.head()
```

```python
y = X["price"]
X = X.drop("price", axis=1)
```

```python
X.head()
```

```python
y.head()
```

```python
y.shape
```

```python
drop_columns = ["id", "made", "numPrevOwners"]
X = X.drop(drop_columns, axis=1)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=0)
```

```python
from xgboost import XGBRegressor

model = XGBRegressor(n_estimators=300, learning_rate=0.05, random_state=42)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error

preds = model.predict(X_valid)
mae = mean_absolute_error(preds, y_valid)

print("MAE:", mae)
```

```python
X_test = pd.read_csv("sample_data/test.csv")
X_test.head()
```

```python
ids = X_test["id"]
X_test = X_test.drop(drop_columns, axis=1)
```

```python
final_preds = model.predict(X_test)
```

```python
output = pd.DataFrame({
    "id": ids,
    "price": final_preds
})

output
```

```python
output.to_csv("sample_data/submission.csv", index=False)
```

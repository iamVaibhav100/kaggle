```python
import numpy as np
import pandas as pd
```

```python
X = pd.read_csv("sample_data/train.csv")
X.head()
```

```python
df = pd.read_csv("sample_data/cubic_zirconia.csv")
df.head()
```

```python
X.shape, df.shape
```

```python
X = pd.concat([X, df])
X.shape
```

```python
X.head()
```

```python
X = X.iloc[:, :-1]
X.head()
```

```python
X.shape
```

```python
X = X.dropna()
```

```python
X.shape
```

```python
drop_columns = ["id", "price"]
y = X["price"]
X = X.drop(drop_columns, axis=1)
```

```python
y.head(5)
```

```python
y.shape
```

```python
cut_mapping = {'Fair': 1, 'Good': 2, 'Very Good': 3, 'Premium': 4, 'Ideal': 5}
color_mapping = {'J': 1, 'I': 2, 'H': 3, 'G': 4, 'F': 5, 'E': 6, 'D': 7}
clarity_mapping = {'I3': 1, 'I2': 2, 'I1': 3, 'SI2': 4, 'SI1': 5, 'VS2': 6, 'VS1': 7, 'VVS2': 8, 'VVS1': 9, 'IF': 10, 'FL': 11}

X['cut'] = X['cut'].map(cut_mapping)
X['color'] = X['color'].map(color_mapping)
X['clarity'] = X['clarity'].map(clarity_mapping)
```

```python
X.head()
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=1)
```

```python
X_train
```

```python
from xgboost import XGBRegressor

model = XGBRegressor(n_estimators=500, learning_rate=0.05, random_state=1)
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error

valid_preds = model.predict(X_valid)
mae = mean_absolute_error(valid_preds, y_valid)
```

```python
mae
```
